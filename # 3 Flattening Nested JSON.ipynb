```python
import pandas as pd
import requests # used to send HTTPS requests to web API -- without this you cannot fetch data from URL. 
import json
```

```python
keyword = input("Please enter your keyword or category")

api_key = '5869a08a34cc47b6b34af2d39d63eaee'

api_url = f'https://newsapi.org/v2/everything?q={keyword}&apiKey={api_key}'

response = requests.get(api_url)
print(response)
```

```python
rawdata = response.json()
print(rawdata)
```

```python
df_normal = pd.json_normalize(rawdata,'articles',sep='_')[['source_name','author','title','description','publishedAt']]
df_normal
```

```python
prettydata = json.dumps(rawdata, indent =5)
print(prettydata)
```

```python
data = [] 

for prettydata in rawdata["articles"]:
    data.append({
        "Source Name" : prettydata["source"]["name"],
        "Author Name" : prettydata["author"],
        "Title" : prettydata["title"],
        "Description" : prettydata["description"],
        "Published Time" : prettydata["publishedAt"]
    })
```

```python
print(data)
```

```python
df = pd.DataFrame(data)
```
