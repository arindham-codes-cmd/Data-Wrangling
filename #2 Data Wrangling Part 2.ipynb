```python
# !pip install rapidfuzz
from rapidfuzz import fuzz
```

```python
def find_near_duplicates(df_ct, column, threshold = 70):
    duplicates =[]
    for i in range (len(df_ct)):
        for j in range(i+1, len(df_ct)): 
            similarity = fuzz.ratio(df_ct[column][i],df_ct[column][j])
            if similarity >= threshold:
                duplicates.append((df_ct['TicketID'][i],df_ct['TicketID'][j], similarity))
    
    return duplicates
```

```python
near_duplicates = find_near_duplicates(df_ct,'CustomerID',threshold = 80)
print("Near duplicates from data set:")
for dup in near_duplicates: 
    print(f'Ticket {dup[0]} and Ticket {dup[1]} are {round(dup[2],2)}% similar.')
```

```python
# !pip install recordlinkage
import recordlinkage
```

```python
# step 1: set up an indexer 
indexer = recordlinkage.Index()
indexer.full() 
links = indexer.index(df_ct)
```

```python
# compare records based on similarity 
compare = recordlinkage.Compare()

# compare using string similarity
compare.string('Contact_Name','Contact_Name',method ='jarowinkler', threshold = 0.7, label = 'name_similarity_JW')

compare.string('Contact ','Contact ',method ='levenshtein', threshold = 0.7, label = 'email_similarity_LS')

compare.exact('Contact ','Contact ',label = 'contact_similarity')

features = compare.compute(links,df_ct)

potential_duplicates = features[(features.sum(axis=1)>=2)]
print(potential_duplicates)
```
